Rerun the code below for new model

```python
## In this cell importing train_test_split module. I will be dividing it to 70-30 - "golden rule".

from sklearn.model_selection import train_test_split

## I'm dropping everything that isn't a numeric
## x is the training attributes, without the class label
x = dataset.drop(['workclass','education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'class'], axis = 1)
x.head()
```

```python
## y is the class label
y = dataset["class"]
y.head()
```

```python
## splitting data to get 70-30
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

```python
## I decided to use a randomForestClassifier as it is an ensemble, and will outperform a lot of other classifiers as it uses the bagging technique,
## by resampling (picking with replacement) and using a system that is similar to a majority voting system. 
from sklearn.ensemble import RandomForestClassifier

## From original data, training on original dataset
model = RandomForestClassifier()

## For Testing data, training on testing dataset
model1 = RandomForestClassifier()
```

```python
## Training the model using instances from original dataset.
model.fit(train_x, train_y)
```

```python
## Prediction of class labels occurs with the line below
predicted = model.predict(test_x)
```

```python
## Only four points - either someone belongs in >=50k or not.
plt.scatter(test_y, predicted)
```

```python
## 81.06% accuracy - pretty good. Look up converting education to a numeric form and this could potentially increase accuracy
model.score(test_x, test_y)
```

```python
## Importing test set - all cells below perform pretty much the same thing. However it only performs on the test set. 
dataset1 = pd.read_csv('adult.test', delimiter = ",", skiprows = 1, index_col = False, names = ["age", "workclass", "final_weight", "education", "education_num", "marital_status", "occupation", "relationship", "race", "sex", "capital_gain", "capital_loss", "hours_per_week", "native_country", "class"])
```

```python
a = dataset1.drop(['workclass','education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'class'], axis = 1)
a.head()
```

```python
b = dataset1["class"]
b.head()
```

```python
train_a, test_a, train_b, test_b = train_test_split(a, b, test_size = 0.3)

model1.fit(train_a, train_b)
```

```python
predicted1 = model.predict(test_a)
```

```python
plt.scatter(test_b, predicted1)
```

```python
model1.score(test_a, test_b)
```

Return around 80 - 82% accuracy for both the original and test set. This changes as I rerun the data. 

```python
## Performing cross validation to evaluate how good the model is performing given this dataset. Model - refers to origin dataset. Model1 - refers to testing dataset.
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, test_x, test_y, cv = 5)
scores1 = cross_val_score(model1, test_a, test_b, cv = 5)
```

```python
print(scores, "Scores for Original dataset")
print(scores1, "Scores for Testing dataset")
```

```python
## This provides the mean score and 95% confidence interval of the score estimate. Observed that the accuracy stays around 80-82%

print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
print("Accuracy: %0.2f (+/- %0.2f)" % (scores1.mean(), scores1.std() * 2))                                       
```
